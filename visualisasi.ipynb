```python
import matplotlib.pyplot as plt 
import numpy as np 

kelas = ('Kelas_25','Kelas_26','Kelas_27','Kelas_28')
jumlah_mhs = (19,26,30,22)
x_coordinat = np.arange(len(kelas))

plt.bar(x_coordinat,jumlah_mhs)
plt.show()
```

```python
plt.bar(x_coordinat,jumlah_mhs,tick_label=kelas)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(x_coordinat,jumlah_mhs,tick_label=kelas)
plt.xticks(rotation=90)
plt.xlabel('Nama Kelas')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Sebaran Kelas Mata Kuliah Data Mining')
plt.show()
```

```python
import pandas as pd 

df = pd.DataFrame({
    'Kelas':kelas,
    'Jumlah_Mhs':jumlah_mhs,
})
df.sort_values(by='Jumlah_Mhs',inplace=True)
x_coordinat = np.arange(len(df))
plt.bar(x_coordinat,df['Jumlah_Mhs'],tick_label=df['Kelas'])
plt.xticks(rotation=90)
plt.xlabel('Nama Kelas')
plt.ylabel('Jumlah')
plt.show()
```
